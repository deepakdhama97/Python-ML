# The Boston Housing dataset is a well-known dataset in the field of machine learning and statistical analysis. It contains information about various housing features in different suburbs of Boston.

```python
# Import the Pandas Library
import pandas as pd
boston=pd.read_csv('Boston1.csv')
```

```python
boston.head()
```

```python
boston.shape
```

The dataset consists of 506 observations with 14 features.
The variables are:

crim : per capita crime rate by town
zn : proportion of residential land zoned for lots over 25,000 sq.ft.
indus : proportion of non-retail business acres per town
chas : Charles River dummy variable (= 1 if tract bounds river; 0 otherwise)
nox : nitric oxides concentration (parts per 10 million)
rm : average number of rooms per dwelling
age : proportion of owner-occupied units built prior to 1940
dis : weighted distances to five Boston employment centers
rad : index of accessibility to radial highways
tax : full-value property-tax rate per $10,000
ptratio : pupil-teacher ratio by town
black : 1000(Bk - 0.63)^2 where Bk is the proportion of blacks by town
lstat : lower status of the population (percent)
medv : median value of owner-occupied homes in $1000s
    
The target variable in this dataset is "medv," which is the median value of owner-occupied homes in $1000s. The other variables represent various features of the suburbs that may impact the median home value.

This dataset is often used for regression analysis and predicting the median value of owner-occupied homes in Boston based on various other variables.

      Preparing Data

```python
x=pd.DataFrame(boston1['lstat'])     # predicted variable
y=pd.DataFrame(boston1['medv'])     # target variable or dependent variable
```

```python
x.size, y.size
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=.2,random_state=1)
```

```python
print(x_train.shape)
print(x_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
from sklearn.linear_model import LinearRegression
regressor= LinearRegression()
regressor.fit(x_train,y_train)            # to train the model
```

```python
print(regressor.intercept_)         # in y=mx+c    m=34.33 i.e --> regressor.intercept_ 
print(regressor.coef_)              # c= -0.9244   i.e -->  regressor.coef_
                                 # medv = -0.9244 and lstat is 34.3349
```

```python
y_pred = regressor.predict(x_test)
y_pred = pd.DataFrame(y_pred, columns=['Predicted'])
```

```python
y_pred
```

```python
y_test
```

```python
from sklearn import metrics
import numpy as np
print('LSTAT Variable')
print('Mean absolute Error: ', metrics.mean_absolute_error(y_test,y_pred))
print('Mean Squared Error: ', metrics.mean_squared_error(y_test,y_pred))
print('Root Mean Squared Error: ', np.sqrt(metrics.mean_squared_error(y_test,y_pred)))
```

  for RM model

```python
x1=pd.DataFrame(boston['rm'])  
y1=pd.DataFrame(boston['medv']) 

from sklearn.model_selection import train_test_split
x1_train, x1_test, y1_train, y1_test = train_test_split(x1,y1,test_size=.2,random_state=1)

from sklearn.linear_model import LinearRegression
regressor= LinearRegression()
regressor.fit(x1_train,y1_train)

print('c:  ',regressor.intercept_)      
print('m:  ',regressor.coef_)  

y1_pred = regressor.predict(x1_test)
y1_pred = pd.DataFrame(y1_pred, columns=['Predicted'])

from sklearn import metrics
import numpy as np
print('RM Variable')
print('Mean absolute Error: ', metrics.mean_absolute_error(y1_test,y1_pred))
print('Mean Squared Error: ', metrics.mean_squared_error(y1_test,y1_pred))
print('Root Mean Squared Error: ', np.sqrt(metrics.mean_squared_error(y1_test,y1_pred)))
```

Mean Absolute Error = MAE= average of  modulus of error  OR average of mod|error|

Mean Squared Error = MSE = average of square of Error
          
Root Mean Squared Error= RMSE = root of MSE




conclusion :  RM model is better than LSTAT

                 For Multiple Variable

```python
x2=pd.DataFrame(boston.loc[:,['rm','lstat']])  
y2=pd.DataFrame(boston['medv']) 

print('x2.size : ', x2.size, '  y2.size: ',  y2.size)

from sklearn.model_selection import train_test_split
x2_train, x2_test, y2_train, y2_test = train_test_split(x2,y2,test_size=.2,random_state=1)

print(x2_train.shape)
print(x2_test.shape)
print(y2_train.shape)
print(y2_test.shape)

from sklearn.linear_model import LinearRegression
regressor= LinearRegression()
regressor.fit(x2_train,y2_train)

print('c:  ',regressor.intercept_)      
print('m:  ',regressor.coef_)  

y2_pred = regressor.predict(x2_test)
y2_pred = pd.DataFrame(y2_pred, columns=['Predicted'])

from sklearn import metrics
import numpy as np
print('RM and LSTAT Variable')
print('Mean absolute Error: ', metrics.mean_absolute_error(y2_test,y2_pred))
print('Mean Squared Error: ', metrics.mean_squared_error(y2_test,y2_pred))
print('Root Mean Squared Error: ', np.sqrt(metrics.mean_squared_error(y2_test,y2_pred)))
```

               Multiple Linear Regression

```python
x2=pd.DataFrame(boston.iloc[:,0:-1])  
y2=pd.DataFrame(boston['medv']) 

print('x2.size : ', x2.size, '  y2.size: ',  y2.size)

from sklearn.model_selection import train_test_split
x2_train, x2_test, y2_train, y2_test = train_test_split(x2,y2,test_size=.2,random_state=5)

print(x2_train.shape)
print(x2_test.shape)
print(y2_train.shape)
print(y2_test.shape)

from sklearn.linear_model import LinearRegression
regressor= LinearRegression()
regressor.fit(x2_train,y2_train)

v=pd.DataFrame(regressor.coef_,index=['co-efficient']).transpose()
w=pd.DataFrame(x2.columns, columns=['Attribute'])

coeff_df = pd.concat([w,v], axis=1, join = 'inner')
print('m: ',coeff_df)

print('c:  ',regressor.intercept_)      


y2_pred = regressor.predict(x2_test)
y2_pred = pd.DataFrame(y2_pred, columns=['Predicted'])

from sklearn import metrics
import numpy as np
print('Multiple Linear Regression')
print('Mean absolute Error: ', metrics.mean_absolute_error(y2_test,y2_pred))
print('Mean Squared Error: ', metrics.mean_squared_error(y2_test,y2_pred))
print('Root Mean Squared Error: ', np.sqrt(metrics.mean_squared_error(y2_test,y2_pred)))
```
