```python
import pandas as pd
import numpy as np
import seaborn as sns
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score
from matplotlib import pyplot as plt
```

<img src="https://miro.medium.com/max/1400/1*cUcY9jSBHFMqCmX-fp8BvQ.jpeg">

```python
X= np.random.rand(50,2)
Y= 2 + np.random.rand(50,2)

Z= np.concatenate((X,Y))
Z=pd.DataFrame(Z) #converting into data frame for ease
```

```python
Z
```

```python
sns.scatterplot(x = Z[0], y = Z[1]);
```

```python
KMean= KMeans(n_clusters=2)       # n_clusters is how many cluster eg. 2,3,4,5,6,7..
KMean.fit(Z)
label=KMean.predict(Z)
```

```python
pd.Series(label).value_counts()
```

```python
print(f'Silhouette Score(n=2): {silhouette_score(Z, label)}')
```

```python
sns.scatterplot(x = Z[0], y = Z[1],hue=label);
```

```python
KMean= KMeans(n_clusters=3)
KMean.fit(Z)
label=KMean.predict(Z)
print(f'Silhouette Score(n=3): {silhouette_score(Z, label)}')
sns.scatterplot(x = Z[0], y = Z[1],hue=label);
```

```python
KMean= KMeans(n_clusters=4)
KMean.fit(Z)
label=KMean.predict(Z)
print(f'Silhouette Score(n=4): {silhouette_score(Z, label)}')
sns.scatterplot(x = Z[0], y = Z[1],hue=label);
```

The optimal cluster in this data is 3 cluster because Silhouette score is max when cluster is 3 

```python
df = sns.load_dataset("iris")
df.head()
```

```python
x = df.iloc[:, :4]
y = df.iloc[:, -1]
```

```python
model = KMeans(n_clusters = 3)
model.fit_predict(x)
```

```python
score = silhouette_score(x, model.labels_)
print("Silhouette Score", round(score, 2))
```

```python
model = KMeans(n_clusters = 4)
model.fit_predict(x)
```

```python
score = silhouette_score(x, model.labels_)
print("Silhouette Score", round(score, 2))
```

```python
model = KMeans(n_clusters = 2)
model.fit_predict(x)
```

```python
score = silhouette_score(x, model.labels_)
print("Silhouette Score", round(score, 2))
```
